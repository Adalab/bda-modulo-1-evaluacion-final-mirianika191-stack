```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []
        self.ventas_totales = 0.0

    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre.lower():
                producto['cantidad'] += cantidad
                print(f"Producto actualizado: {nombre}, nueva cantidad: {producto['cantidad']}")
                return
        nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
        self.inventario.append(nuevo_producto)
        print(f"Producto agregado: {nombre}, precio: {precio}, cantidad: {cantidad}")

    def ver_inventario(self):
        if not self.inventario:
            print("Inventario vacío.")
        else:
            print("\nInventario actual:")
            for producto in self.inventario:
                print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")

    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre.lower():
                print(f"Producto encontrado - Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
        print(f"Producto '{nombre}' no encontrado en el inventario.")

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto['nombre'].lower() == nombre.lower():
                producto['cantidad'] += cantidad
                print(f"Stock actualizado para {nombre}. Nueva cantidad: {producto['cantidad']}")
                return
        print(f"No se encontró el producto '{nombre}' en el inventario.")

    def eliminar_producto(self, nombre):
        for i, producto in enumerate(self.inventario):
            if producto['nombre'].lower() == nombre.lower():
                self.inventario.pop(i)
                print(f"Producto '{nombre}' eliminado del inventario.")
                return
        print(f"No se encontró el producto '{nombre}' en el inventario.")

    def calcular_valor_inventario(self):
        valor_total = 0
        for producto in self.inventario:
            valor = producto['precio'] * producto['cantidad']
            valor_total += valor
        print(f"\nValor total del inventario: ${valor_total}")
        return valor_total


# --------------------------------------------
# Pruebas / Demostración del funcionamiento
# --------------------------------------------

# Crear instancia de la tienda
mi_tienda = TiendaOnline()

# Agregar productos
mi_tienda.agregar_producto("Camisa", 20, 40)
mi_tienda.agregar_producto("Pantalón", 30, 30)
mi_tienda.agregar_producto("Zapatos", 50, 40)
mi_tienda.agregar_producto("Camisa", 20, 10)  # Actualiza cantidad de Camisas

# Ver inventario
mi_tienda.ver_inventario()

# Buscar un producto existente
mi_tienda.buscar_producto("Camisa")

# Buscar un producto que no existe
mi_tienda.buscar_producto("Gorra")

# Actualizar stock (sumar y restar)
mi_tienda.actualizar_stock("Pantalón", -5)  # Quitar 5 pantalones
mi_tienda.actualizar_stock("Gorra", 10)     # Producto no existe

# Ver inventario actualizado
mi_tienda.ver_inventario()

# Eliminar un producto
mi_tienda.eliminar_producto("Zapatos")
mi_tienda.eliminar_producto("Zapatillas")  # Producto no existe

# Ver inventario tras eliminación
mi_tienda.ver_inventario()

# Calcular valor total del inventario
mi_tienda.calcular_valor_inventario()
```
