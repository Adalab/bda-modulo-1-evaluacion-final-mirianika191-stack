
Verificar si un número es par o impar

Descripción: Dado un número, verifica si es par o impar.

```python
lista_numero_par = [2, 4, 6, 8, 10]
if numero_elegido in lista_numero_par
    print("par")
elif numero_elegido not in lista_numero_par:
    print("impar")
```
